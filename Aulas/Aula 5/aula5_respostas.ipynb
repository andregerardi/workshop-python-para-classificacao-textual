## Como contar o número de ocorrências de determinada palavra no corpus?

Vamos contar quantas vezes as palavras "campanha", "voto" e "São Paulo" ocorreram.

```python
## indica a contagem de citações, que é zero
cont_campanha = 0
cont_voto = 0
cont_sp = 0

## processamento usando a função count()
for frase in frases.texto: #para cada frase do corpus, faça:
  cont_campanha = cont_campanha + frase.lower().count('campanha')
  cont_voto = cont_voto + frase.lower().count('vote')
  cont_sp = cont_sp + frase.lower().count('são paulo')

## total de citações de palavras no texto
print('Campanha: ', cont_campanha)
print('Voto: ', cont_voto)
print('São Paulo: ', cont_sp)
```

## É possível saber se os candidatos estão utilizando seu número de urna para divulgação de campanha?

Vamos testar!

```python
#para cada frase do corpus, faça:
for frase in frases.texto:
  numeros = re.findall('[0-9]{4}', frase)
  print('Números encontrados:', numeros)
```

## Como classificar o material de pesquisa usando Regex?

```python
for frase in frases.texto:
  ## Se a condição abaixo for encontrada, classifica como campanha.
  if re.search(r'(candidat\w+|deputad\w+|vot\w+)', frase.lower()):         # Se contem as plavras indicadas
      classe = 'campanha'
      print(classe,'-',frase)
  elif re.search(r'cont\w+ com (o seu apoio|a sua ajuda)', frase.lower()): # Se contém orações ou frases espefícias
      classe = 'campanha'
      print(classe,'-',frase)
  elif re.findall('[0-9]{4}', frase.lower()):                             # Se contém números com formatos específicos
      classe = 'campanha'
      print(classe,'-',frase)
  ## Se nenhuma condição acima for encontrada, classifica como não campanha.
  else:
    classe = 'não campanha'
    print(classe,'-',frase)
```

# **Modelo usando base de dados**

```python
# Obs: Esse modelo demora aproximadamente 8 minutos para classificar as 30 mil linhas

## Bibliotecas
import matplotlib.pyplot as plt
from tqdm import tqdm
import pandas as pd
import re

## Importa base de dados
from google.colab import drive
drive.mount('/content/gdrive')

df = pd.read_excel(r'/content/gdrive/MyDrive/(2026) workshop python para cientistas politicos/workshop-python-para-classificacao-textual/Bases de dados/lista_juncao_posts_deputados(nomes_datas_ordenadas)_2022.xlsx')

## Modelo de classificação usando o looping

# Cria uma lista vazia para salvar os valores
lista_classificacao_looping = []

# Lopping da clasificação
# lembre-se, usamos os dados da coluna 'message' para pegar o texto dos posto e colocar no looping
for frase in tqdm(df.message):
  try:
    if re.search(r'(candidat\w+|deputad\w+|vot\w+)', frase.lower()):
        classe = 'campanha'
    elif re.search(r'cont\w+ com (o seu apoio|a sua ajuda)', frase.lower()):
        classe = 'campanha'
    elif re.findall(r'[0-9]{4}', frase.lower()):
        classe = 'campanha'
    else:
      classe = 'não campanha'
  except:
    classe = 'erro'

  ## a função append salva o valores na lista vazia a cada iteração
  lista_classificacao_looping.append([frase,classe])

  ## converte os dados salvos numa tabela
  lista_resutados = pd.DataFrame(lista_classificacao_looping, columns=['Frase', 'Classificacao'])

  ## salva a tabela como um arquivo Excel
  ## salvaremos na pasta da aula 5
  lista_resutados.to_excel(r"/content/gdrive/MyDrive/(2026) workshop python para cientistas politicos/workshop-python-para-classificacao-textual/Aulas/Aula 5/lista_classificacao_looping_workshop.xlsx")
```
