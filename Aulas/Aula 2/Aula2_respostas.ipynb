# **Métodos exclusivos para trabalhar com String**

Além das operações matemáticas, existem algumas que são exclusivas de string e podem ser muito úteis para a análise linguística.

````python
help(str)
````

Para aplicar métodos exclusivos de string, você precisa colocar a string, seguida de ponto e então chamar o método. <br>Por exemplo, string.lower().

Alguns métodos:

* lower() - transforma todas as letras em minúsculo
* upper() - transforma todas as letras em maísculo
* islower() - verifica se a palavra está em minúsculo
* isupper() - verifica se a palavra está em maísculo

````python
frase = 'Inteligência artificial não é consciente'
````

As funções **lower()** e **upper()** podem ser usadas para normalizar as palavras.

````python
frase.lower()
````

````python
frase.upper()
````

````python
frase.title()
````

````python
# função para verificar a ingualdade do conteúdo de variáveis. Nestes casos a frase maiúscula é diferente de minúscula
frase.upper() == frase.lower()
````

As funções **islower()** e **isupper()** retornam um valor que é do tipo booleano. Um booleano só pode ter os valores: True (caso satistaça um critério) ou False (caso contrário).

````python
frase.islower()
````

````python
palavra_minuscula = frase.lower()
palavra_minuscula
````

````python
minuscula_valor = palavra_minuscula.islower()
minuscula_valor
````

````python
type(minuscula_valor)
````
